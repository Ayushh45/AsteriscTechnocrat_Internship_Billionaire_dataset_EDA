```python
plt.figure(figsize=(20,10))

sbn.histplot(x='Age',data=df)
plt.show()
```

```python
a = df['Industry'].value_counts().head()
print(a)
Index=a.index
print(Index)
Values=a.values
print(Values)
```

```python
plt.figure(figsize=(5,5))
plt.pie(Values,labels=Index)
plt.title('5 category',fontsize=20)
plt.show()
```

```python
b = df['Country'].value_counts().head()
b
```

```python
Index1 = b.index
Index1
```

```python
Value1 = b.values
Value1
```

```python
plt.figure(figsize=(5,5))
plt.pie(Value1,labels=Index1,colors=['blue','maroon','red','violet','black'])
plt.title('5 Country',fontsize=20)
plt.show()
```

```python
c = df['Source'].value_counts().head()
c
Index2 = c.index
Index2
Value2 = c.values
Value2
plt.figure(figsize=(8,5))
plt.bar(Index2,Value2,color='#B266FF')
plt.title('Source',fontsize=20,color='#6600CC')
plt.show()
```

```python
d = df['Industry'].value_counts().head()
d
Index3 = d.index
Index3
Value3 = d.values
Value3
plt.figure(figsize=(10,5))
plt.bar(Index3,Value3)
plt.title('Industry',fontsize=20)
plt.show()
```

```python
e = df['Age'].value_counts().head()
e
Index4 = e.index
Index4
Value4 = e.values
Value4
plt.figure(figsize=(10,5))
plt.bar(Index4,Value4,color=['blue','maroon','red','violet','black'])
plt.title('Most Billionaires Age Category',fontsize=20)
plt.show()
```
